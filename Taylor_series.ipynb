```python

import math

ITERATIONS = 5

def my_cosh(x):
    """
    Вычисление гиперболического косинуса при помощи частичного суммирования
    ряда Тейлора для окрестности 0
    """
    x_pow = 1
    multiplier = 1
    partial_sum = 1
    for n in range(1, ITERATIONS):
        x_pow *= x**2 
        multiplier *= 1 / (2 * n - 1) / (2 * n)
        partial_sum += x_pow * multiplier
    
    return partial_sum

help(math.cosh)
help(my_cosh)

print(math.cosh(0.3))
print(my_cosh(0.3))

import matplotlib.pyplot as plt
import numpy as np

vs = np.vectorize(my_cosh)
print(my_cosh, vs)

angles = np.r_[-7:7:0.001]

plt.style.use("seaborn-darkgrid")
fig, axes = plt.subplots ()

axes.set_xlabel ("Ось X")
axes.set_ylabel ("Ось Y")


plt.plot(angles, np.cosh(angles), color="red", linestyle="-", label="math.cosh")
plt.plot(angles, vs(angles), color="blue", linestyle="-", label="my_cosh")
axes.legend (loc="center")

plt.show()
```
